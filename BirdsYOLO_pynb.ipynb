```python
import cv2
import os

# Load the video
video_path = '/content/video.mp4'
cap = cv2.VideoCapture(video_path)

if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
    exit()

# Get video properties
fps = cap.get(cv2.CAP_PROP_FPS)
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # Codec for .mp4

# Define output video path
output_video_path = '/content/video_birds_detailed.mp4'
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

# Weight Estimation Constant
CALIBRATION_CONSTANT_K = 0.00005 # Heuristic: Weight = K * (Area)^1.5

# Perform prediction in stream mode, filtering for bird class (ID 14)
# The `classes=[14]` argument ensures that only detections for 'bird' are returned.
results_generator = model.predict(source=video_path, stream=True, classes=[14])

for r in results_generator:
    # Get the original frame (numpy array)
    frame = r.orig_img

    # Count the number of bird detections in the current frame
    bird_count = len(r.boxes)

    # Overlay the total bird count on the frame
    text_total_count = f"Total Birds: {bird_count}"
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale_total = 1
    font_thickness_total = 2
    text_color_total = (0, 0, 255) # Red color for total count
    text_position_total = (50, 50) # Top-left corner for total count

    cv2.putText(frame, text_total_count, text_position_total, font, font_scale_total, text_color_total, font_thickness_total, cv2.LINE_AA)

    # Draw bounding boxes, confidence, and estimated weight for each detected bird
    for i, box in enumerate(r.boxes.xywh):
        x_center, y_center, w_box, h_box = map(float, box)
        x1, y1, x2, y2 = map(int, r.boxes.xyxy[i]) # Get xyxy for drawing the rectangle

        # Calculate area and estimated weight
        area = w_box * h_box
        estimated_weight = CALIBRATION_CONSTANT_K * (area**1.5)

        # Get confidence score
        confidence = r.boxes.conf[i]

        # Draw bounding box
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2) # Green color, 2 pixels thick

        # Overlay confidence and weight for each bird
        info_text = f"C:{confidence:.2f}|W:{estimated_weight:.2f}g"
        font_scale_individual = 0.5
        font_thickness_individual = 1
        text_color_individual = (255, 255, 255) # White text for individual bird info

        # Position text above the bounding box
        text_origin = (x1, y1 - 10 if y1 - 10 > 10 else y1 + 20) # Avoid going off screen
        cv2.putText(frame, info_text, text_origin, font, font_scale_individual, text_color_individual, font_thickness_individual, cv2.LINE_AA)

    # Write the modified frame to the output video
    out.write(frame)

# Release video capture and writer objects
cap.release()
out.release()

print(f"Processed video saved to {output_video_path}")
```
